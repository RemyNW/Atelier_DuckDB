```python
import duckdb
```

```python
conn = duckdb.connect('parquet_directory/data_lake.duckdb')

result = conn.execute("SELECT COUNT(*) FROM information_schema.tables").fetchall()
print("Nombre de tables dans la base de données :", result[0][0])
```

```python
conn = duckdb.connect('parquet_directory/data_lake.duckdb')
result = conn.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'raw'").fetchall()

print("Tables dans la base de données :")
for table in result:
    print(table[0])

# Fermez la connexion
conn.close()
```

```python
conn = duckdb.connect('parquet_directory/data_lake.duckdb')
result = conn.execute("SELECT * FROM raw.branches").fetchall()

for row in result:
    branch, commit_info, status = row
    print(f"Branch: {branch}")
    print(f"Commit SHA: {commit_info['sha']}")
    print(f"Commit URL: {commit_info['url']}")
    print(f"Status: {status}")
    print('-' * 40)
```

```python
result = conn.execute("SELECT * FROM raw.issues").fetchall()
print(result)
conn.close()
```
